This analysis uses a public dataset in BigQuery called London Bicycle Hires. 

To create tables the steps are to first create a project, then create a database, a dataset, then tables. Since the data is already publicly available, I can immediately start analyzing the data.

The dataset has two tables called cycle_hires and cycle_stations. Cycle_hires has 14 fields and has 83M observations. Dataset compose of a mixture of time, date, strings, floats, as well as boolean values. 

# NO OF BICYCLE HIRES LONGER THAN 20 MINS

To do this, I just identified and counted all rows with length of rental >= 1200 seconds.


```
SELECT
  COUNT (*) AS no_of_bike_trips
FROM
  `bigquery-public-data.london_bicycles.cycle_hire`
WHERE
  duration >=1200
```

RESULTS:

```
no_of_bike_trips
26441016

```






#What are the names of the stations that bike_id 1710 started from?

```
SELECT
  DISTINCT start_station_name,
FROM
  `bigquery-public-data.london_bicycles.cycle_hire`
WHERE
  bike_id = 1710
```
RESULTS:

```
798 rows of addresses
```




#How many bike_ids have ended at "Moor Street, Soho"?

```
SELECT
  COUNT (DISTINCT bike_id) AS no_of_bikes,
FROM
  `bigquery-public-data.london_bicycles.cycle_hire`
WHERE
  end_station_name = 'Moor Street, Soho'
```

RESULTS: 

```
no_of_bikes
24998

```



#What is the station_id for "Canton Street, Poplar"?

```
SELECT
  id AS station_id
FROM
  `bigquery-public-data.london_bicycles.cycle_stations`
WHERE
  name = "Canton Street, Poplar"
```
RESULTS:


```
station_id	
487
```




#What is the name of the station whose ID is 111?

```
SELECT
  DISTINCT start_station_name AS station_name
FROM
  `bigquery-public-data.london_bicycles.cycle_hire`
WHERE
  start_station_id = 111
```
RESULTS:

```
station_name	
Park Lane , Hyde Park
```





#How many distinct bike_ids had trip durations greater than 2400 seconds (or 40 minutes)?



```
SELECT
  COUNT (DISTINCT bike_id) AS no_of_bike_trips
FROM
  `bigquery-public-data.london_bicycles.cycle_hire`
WHERE
  duration > 2400
```

RESULTS:


```
no_of_bike_trips
31636
```
